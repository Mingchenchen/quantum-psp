```python
# (C) Copyright Renata Wong 2020.

from qiskit import *

w = QuantumRegister(2,'w')
anc = QuantumRegister(3,'anc')
c = QuantumRegister(1,'c')
a = QuantumRegister(2,'a')
x = QuantumRegister(2,'x')
qc = QuantumCircuit(w,anc,c,a,x)

qc.ccx(w[0],w[1],anc[0])

qc.ccx(anc[0],a[0],anc[1])
# first two gates
qc.cx(anc[1],x[0])
qc.cx(anc[1],c[0])
# uncompute 
qc.ccx(anc[0],a[0],anc[1])
# continue
qc.ccx(anc[0],a[1],anc[1])
# third gate
qc.ccx(anc[1],x[1],anc[2])
qc.cx(anc[2],c[0])
# uncompute
qc.ccx(anc[1],x[1],anc[2])
qc.ccx(anc[0],a[1],anc[1])
# continue, gate 4
qc.ccx(anc[0],c[0],anc[1])
qc.ccx(anc[1],x[0],anc[2])
qc.cx(anc[2],a[0])
# uncompute
qc.ccx(anc[1],x[0],anc[2])
qc.ccx(anc[0],c[0],anc[1])
# continue gate 5
qc.cx(anc[0],x[0])
# gate 6
qc.ccx(anc[0],c[0],anc[1])
qc.cx(anc[1],x[0])
# uncompute
qc.ccx(anc[0],c[0],anc[1])
# continue gate 7
qc.ccx(anc[0],c[0],anc[1])
qc.ccx(anc[1],x[0],anc[2])
qc.cx(anc[2],a[0])
# uncompute
qc.ccx(anc[1],x[0],anc[2])
qc.ccx(anc[0],c[0],anc[1])
# continue, gate 8
qc.cx(anc[0],x[0])
# gate 9
qc.ccx(anc[0],a[1],anc[1])
qc.ccx(anc[1],x[1],anc[2])
qc.cx(anc[2],c[0])
# uncompute
qc.ccx(anc[1],x[1],anc[2])
qc.ccx(anc[0],a[1],anc[1])
# continue, gate 10
qc.ccx(anc[0],a[0],anc[1])
qc.cx(anc[1],c[0])
# uncompute
qc.ccx(anc[0],a[0],anc[1])
# continue, gate 11
qc.ccx(anc[0],a[0],anc[1])
qc.cx(anc[1],x[0])
# uncompute
qc.ccx(anc[0],a[0],anc[1])
# continue, gate 12
qc.ccx(anc[0],a[1],anc[1])
qc.cx(anc[1],x[1])
# uncompute
qc.ccx(anc[0],a[1],anc[1])

qc.ccx(w[0],w[1],anc[0])

qc.draw(output='latex',filename="test.png", scale=1,plot_barriers=False)
```
